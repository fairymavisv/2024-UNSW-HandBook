```python
import requests
import json
import sys
import os
from bs4 import BeautifulSoup
```

```python
header = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "Accept-Language": "zh-CN,zh;q=0.9",
    "Sec-Ch-Ua": '"Not_A Brand";v="8", "Chromium";v="120", "Microsoft Edge";v="120"',
    "Sec-Ch-Ua-Mobile": "?0",
    "Sec-Ch-Ua-Platform": "Windows",
    "Sec-Fetch-Dest": "document",
    "Sec-Fetch-Mode": "navigate",
    "Sec-Fetch-Site": "none",
    "Sec-Fetch-User": "?1",
    "Upgrade-Insecure-Requests": "1",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36 Edg/120.0.0.0",
}
timetable_url = "https://timetable.unsw.edu.au/"
session = requests.Session()

```

```python
response = session.get(timetable_url+"2024/COMPKENS.html", headers=header, stream=False)
content = response.content
soup = BeautifulSoup(content, "html.parser")
```

```python
form_body = soup.find('td', {"class" : "classSearchFormBody"} )
degrees = form_body.find_all('td', {"class": "classSearchMinorHeading"})
ug_course = degrees[0].find_parent('tr').find_next('tr')
pg_course = degrees[2].find_parent('tr').find_next('tr')
```

```python
ug_courses = set()
for each in ug_course.findAll('tr', {"class" : "rowLowlight"}):
    code = each.findNext('td', {"class" : "data"})
    name = code.findNext('td', {"class" : "data"})
    uoc = name.findNext('td', {"class" : "data"})
    ug_courses.add(code.text.strip())
    with open(f"data/courses/{code.text}.json", "w") as f:
        json.dump({
            "code": code.text,
            "name": name.text,
            "uoc": uoc.text,
            "ug": True,
            "pg": False,
            "handbook": f"https://www.handbook.unsw.edu.au/undergraduate/courses/2024/{code.text}",
            "timetable": f"https://timetable.unsw.edu.au/2024/{code.text}"
        }, f)
for each in ug_course.findAll('tr', {"class" : "rowHighlight"}):
    code = each.findNext('td', {"class" : "data"})
    name = code.findNext('td', {"class" : "data"})
    uoc = name.findNext('td', {"class" : "data"})
    ug_courses.add(code.text.strip())
    with open(f"data/courses/{code.text}.json", "w") as f:
        json.dump({
            "code": code.text,
            "name": name.text,
            "uoc": uoc.text,
            "ug": True,
            "pg": False,
            "handbook": f"https://www.handbook.unsw.edu.au/undergraduate/courses/2024/{code.text}",
            "timetable": f"https://timetable.unsw.edu.au/2024/{code.text}"
        }, f)
print(f"Total {len(ug_courses)} ug_courses", ug_courses)
pg_courses = set()
for each in pg_course.findAll('tr', {"class" : "rowLowlight"}):
    code = each.findNext('td', {"class" : "data"})
    name = code.findNext('td', {"class" : "data"})
    uoc = name.findNext('td', {"class" : "data"})
    pg_courses.add(code.text.strip())
    if code.text.strip() in ug_courses:
        print(f"Duplicate course {code.text}")
        with open(f"data/courses/{code.text}.json", "w") as f:
            json.dump({
                "code": code.text,
                "name": name.text,
                "uoc": uoc.text,
                "ug": True,
                "pg": True,
                "handbook": f"https://www.handbook.unsw.edu.au/undergraduate/courses/2024/{code.text}",
                "timetable": f"https://timetable.unsw.edu.au/2024/{code.text}"
            }, f)
    else:
        with open(f"data/courses/{code.text}.json", "w") as f:
            json.dump({
                "code": code.text,
                "name": name.text,
                "uoc": uoc.text,
                "ug": False,
                "pg": True,
                "handbook": f"https://www.handbook.unsw.edu.au/postgraduate/courses/2024/{code.text}",
                "timetable": f"https://timetable.unsw.edu.au/2024/{code.text}"
            }, f)
for each in pg_course.findAll('tr', {"class" : "rowHighlight"}):
    code = each.findNext('td', {"class" : "data"})
    name = code.findNext('td', {"class" : "data"})
    uoc = name.findNext('td', {"class" : "data"})
    pg_courses.add(code.text.strip())
    if code.text.strip() in ug_courses:
        print(f"Duplicate course {code.text}")
        with open(f"data/courses/{code.text}.json", "w") as f:
            json.dump({
                "code": code.text,
                "name": name.text,
                "uoc": uoc.text,
                "ug": True,
                "pg": True,
                "handbook": f"https://www.handbook.unsw.edu.au/undergraduate/courses/2024/{code.text}",
                "timetable": f"https://timetable.unsw.edu.au/2024/{code.text}"
            }, f)
    else:
        with open(f"data/courses/{code.text}.json", "w") as f:
            json.dump({
                "code": code.text,
                "name": name.text,
                "uoc": uoc.text,
                "ug": False,
                "pg": True,
                "handbook": f"https://www.handbook.unsw.edu.au/postgraduate/courses/2024/{code.text}",
                "timetable": f"https://timetable.unsw.edu.au/2024/{code.text}"
            }, f)
print(f"Total {len(pg_courses)} pg_courses", pg_courses)
print(f"Intersection {len(ug_courses.intersection(pg_courses))} courses", ug_courses.intersection(pg_courses))
```
